```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
from sklearn.preprocessing import StandardScaler

def get_test_accuracies_example_style(number):
    df_train = pd.read_csv(f"./monks_data/monks-{number}.train", header=None)

    df_test = pd.read_csv(f"./monks_data/monks-{number}.test", header=None)

    X_train = df_train.iloc[:, 1:].values

    y_train = df_train.iloc[:, 0].values

    X_test = df_test.iloc[:, 1:].values

    y_test = df_test.iloc[:, 0].values

    model_sgd = Sequential()

    model_sgd.add(Dense(8, activation='tanh', input_dim=6))  

    model_sgd.add(Dense(6, activation='relu'))              

    model_sgd.add(Dense(1, activation='sigmoid'))           

    model_sgd.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

    history_sgd = model_sgd.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32)

    model_adam = Sequential()

    model_adam.add(Dense(8, activation='tanh', input_dim=6))  

    model_adam.add(Dense(6, activation='relu'))              

    model_adam.add(Dense(1, activation='sigmoid'))           

    model_adam.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    history_adam = model_adam.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32)

    test_accuracy_sgd = history_sgd.history['val_accuracy']

    test_accuracy_adam = history_adam.history['val_accuracy']

    return test_accuracy_sgd, test_accuracy_adam

test_accuracy_sgd, test_accuracy_adam = get_test_accuracies_example_style(1)

print("SGD Test Accuracy:", test_accuracy_sgd)

print("Adam Test Accuracy:", test_accuracy_adam)
```

```python
import matplotlib.pyplot as plt

def plot_accuracies_for_monks(number):
    test_accuracy_sgd, test_accuracy_adam = get_test_accuracies_example_style(number)

    epochs = range(1, 51)

    plt.figure(figsize=(10, 6))

    plt.plot(epochs, test_accuracy_sgd, label="SGD Optimizer", marker="o", color="blue")

    plt.plot(epochs, test_accuracy_adam, label="Adam Optimizer", marker="s", color="orange")

    plt.xlabel("Epoch")

    plt.ylabel("Test Accuracy")

    plt.title(f"Testing Accuracy vs. Epochs for Monks-{number} Dataset")

    plt.legend()

    plt.grid(True)

    plt.show()

plot_accuracies_for_monks(1)
```

```python
import matplotlib.pyplot as plt

def record_and_plot_test_accuracies(number):

    test_accuracy_sgd, test_accuracy_adam = get_test_accuracies_example_style(number)

    highlighted_epochs = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

    sgd_highlighted = [test_accuracy_sgd[epoch - 1] for epoch in highlighted_epochs]

    adam_highlighted = [test_accuracy_adam[epoch - 1] for epoch in highlighted_epochs]

    plt.figure(figsize=(10, 6))

    plt.plot(highlighted_epochs, sgd_highlighted, label="SGD Optimizer", marker="o", color="blue")

    plt.plot(highlighted_epochs, adam_highlighted, label="Adam Optimizer", marker="s", color="orange")

    plt.xlabel("Epoch")

    plt.ylabel("Test Accuracy")

    plt.title(f"Testing Accuracy vs. Epochs for Monks-{number} Dataset")

    plt.legend()

    plt.grid(True)

    plt.show()

record_and_plot_test_accuracies(1)
```
