```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import combinations

# List of numerical columns to plot
numerical_columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'slope', 'ca']

# Generate scatter plots for each unique pair of numerical features
for feature1, feature2 in combinations(numerical_columns, 2):  # Only unique pairs generated
    plt.figure(figsize=(8, 6))  # Set larger figure size for clarity
    sns.scatterplot(data=data, x=feature1, y=feature2, hue='target', palette='Set1')
    plt.title(f'Scatter Plot: {feature1} vs {feature2}')
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.legend(title='Target')
    plt.show()
```
