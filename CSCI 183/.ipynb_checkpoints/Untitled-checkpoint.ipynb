```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Load the diabetes dataset
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = diabetes.target

# Calculate correlations of each feature with the target
correlations = df.corrwith(pd.Series(y))

# Sort the correlations to find the strongest positive and negative correlations
correlations_sorted = correlations.sort_values(ascending=False)
strong_pos = correlations_sorted.head(2)
strong_neg = correlations_sorted.tail(2)

print("Strong Positive Correlations:\n", strong_pos)
print("Strong Negative Correlations:\n", strong_neg)

# Find features with correlations close to zero (weak associations)
weak_associations = correlations[(correlations > -0.1) & (correlations < 0.1)]
print("Weak Associations:\n", weak_associations)

# Perform Simple Linear Regression and store evaluation results
evaluation_results = []

for feature in df.columns:
    x = df[[feature]]
    
    linear_reg_model = LinearRegression()

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

    linear_reg_model.fit(x_train, y_train)
    
    y_prediction = linear_reg_model.predict(x_test)
    mse = mean_squared_error(y_test, y_prediction)

    coefficient = linear_reg_model.coef_[0]
    intercept = linear_reg_model.intercept_
    
    evaluation_results.append({
        'Feature': feature,
        'Correlation': correlations[feature],
        'MSE': mse,
        'Coefficient': coefficient,
        'Intercept': intercept
    })

# Create a DataFrame with the evaluation results and display it
evaluation_df = pd.DataFrame(evaluation_results)
print(evaluation_df)

# Plot scatter and regression line for each feature
for feature in df.columns:
    x = df[[feature]]
    
    # Fit the model
    linear_reg_model = LinearRegression()
    linear_reg_model.fit(x, y)
    y_prediction = linear_reg_model.predict(x)
    
    # Plot
    plt.scatter(x, y, color='blue', label='Original data')
    plt.plot(x, y_prediction, color='red', label='Regression line')
    plt.title(f"Feature: {feature}")
    plt.xlabel(feature)
    plt.ylabel("Target (Disease Progression)")
    plt.legend()
    plt.show()

# Perform Multiple Linear Regression
x_all = df
linear_reg_model_all = LinearRegression()

x_train_all, x_test_all, y_train_all, y_test_all = train_test_split(x_all, y, test_size=0.2, random_state=42)

linear_reg_model_all.fit(x_train_all, y_train_all)

y_prediction_all = linear_reg_model_all.predict(x_test_all)
mse_all = mean_squared_error(y_test_all, y_prediction_all)

print("MSE for multiple linear regression:", mse_all)
```
